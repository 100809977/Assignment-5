```python
#Load the libraries
import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load all the excel files from the system
dataset = sorted(glob.glob("data*.csv"))
dataset
```

```python
#Build function for visualization

def visualize(filename):
    """Plot the histogram"""
    df = pd.read_csv(filepath_or_buffer = filename)
    fig, axis = plt.subplots(nrows=1, ncols=1, figsize=(15,5))
    plt.plot(df.mean(axis=0), label="Average")
    plt.plot(df.max(axis=0), label='Max')
    plt.plot(df.min(axis=0), label="Min")
    plt.legend(loc="upper right")

    axis.set_title(filename, fontsize=18)
    axis.set_xlabel("Features", fontsize=18)
    axis.set_ylabel("Values", fontsize=18)
    
    matplotlib.pyplot.show()
```

```python
#Build function for issue identification

def identify_issue(filename):
    """identidy issue for all 4 datasets"""
    df=pd.read_csv(filepath_or_buffer = filename)
    
    max_feature0 = df.iloc[:,0].max()

    max_feature20 = df.iloc[:,20].max()

    total = sum(df.min())

    if max_feature0 == 0 and max_feature20 == 20:
        print("I doubt the integrity of this dataset.")
    elif total == 0:
        print("Why is this happening?")
    else:
        print("Dataset looks good!")
```

```python
#Add a for loop to run each of the datasets and plot the line graph.

for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    identify_issue(datasets)
```
